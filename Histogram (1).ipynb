### Write a Python program using OpenCV to perform histogram equalization on the given grayscale image **"parrot.jpg"**.

The program should:

- Import the required libraries (**OpenCV, NumPy, Matplotlib**).
- Read the image **"parrot.jpg"** in **grayscale** format.
- Display the grayscale image and plot its histogram.
- Perform **histogram equalization** on the grayscale image using `cv2.equalizeHist()` to enhance its contrast.
- Using Matplotlib, display the following in a **2 × 2 grid layout**:
  - Original Grayscale Image
  - Histogram of Original Image
  - Enhanced (Equalized) Image
  - Histogram of Enhanced Image


```python
# Import required libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the image in grayscale format
img = cv2.imread("parrot.jpg", 0)

```

```python
# Display the grayscale image.
# Display the grayscale image
plt.imshow(img, cmap='gray')
plt.title("Grayscale Image")
```

```python
# Plot the histogram of the grayscale image
plt.hist(img.ravel(), 256, range=[0, 256])
plt.title("Histogram of Grayscale Image")
plt.show()
```

```python
# Perform histogram equalization
equalized_img = cv2.equalizeHist(img)
```

```python
# Display [1] the Original Image (Gray Image) and its Histogram, and [2] the Enhanced Image and its Histogram using a 2×2 layout in Matplotlib.
# Display Original and Enhanced Images with Histograms

plt.figure(figsize=(10, 8))

# Original Grayscale Image
plt.subplot(221);plt.imshow(img, cmap='gray');plt.title("Original Grayscale Image");plt.axis("off")

# Histogram of Original Image
plt.subplot(222);plt.hist(img.ravel(), 256, [0, 256]);plt.title("Histogram of Original Image");plt.xlabel("Pixel Intensity");plt.ylabel("Frequency")

# Enhanced (Equalized) Image
plt.subplot(223);plt.imshow(equalized_img, cmap='gray');plt.title("Enhanced Equalized Image");plt.axis("off")
# Histogram of Enhanced Image
plt.subplot(224);plt.hist(equalized_img.ravel(), 256, [0, 256]);plt.title("Histogram of Enhanced Image");plt.xlabel("Pixel Intensity");plt.ylabel("Frequency")

plt.tight_layout()
plt.show()


```

### Write a Python program using OpenCV to perform color image enhancement through histogram equalization in the HSV color space on the given image **"parrot.jpg"**.

The program should:

- Import the required libraries (**OpenCV, NumPy, Matplotlib**).
- Read the color image **"parrot.jpg"**.
- Plot the **histogram of the original color image** (use the **B, G, and R** channels).
- Convert the image from **BGR to HSV** color space.
- Apply **histogram equalization** to one specific channel to improve brightness and contrast using `cv2.equalizeHist()`.
- Convert the enhanced HSV image back to **BGR** format.
- Using Matplotlib, display the following in a **2 × 2 grid layout**:
  - Original Color Image
  - Histogram of Original Image (BGR channels)
  - Enhanced Color Image
  - Histogram of Enhanced Image (BGR channels)


```python
# Import required libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt


```

```python
# Read the colorgiven parrot.jpg image.
img = cv2.imread("parrot.jpg")



```

```python
# Plot the histogram of colour image
colors = ['r','g','b']

for i, c in enumerate(colors):
    plt.hist(img[:, :, i].ravel(), bins=256, range=[0, 256], color=c, alpha=0.5)

plt.title("Color Image Histogram")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.show()



```

```python
# Convert BGR image to HSV
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
# Perform histogram equalization on V channel
hsv_img[:, :, 2] = cv2.equalizeHist(hsv_img[:, :, 2])
```

```python
# Convert HSV image back to BGR format
enhanced_img = cv2.cvtColor(hsv_img, cv2.COLOR_HSV2BGR)
```

```python
# Display



plt.figure(figsize=(10, 8))

# Original Color Image
plt.subplot(221);plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB));plt.title("Original Color Image");plt.axis("off")

# Histogram of Original Image
plt.subplot(222)

colors = ['b', 'g', 'r']
for i, c in enumerate(colors):
    plt.hist(img[:, :, i].ravel(), bins=256, range=[0, 256], color=c, alpha=0.5)

plt.title("Histogram of Original Image")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

# Enhanced Color Image
plt.subplot(223);plt.imshow(cv2.cvtColor(enhanced_img, cv2.COLOR_BGR2RGB));plt.title("Enhanced Color Image");plt.axis("off")

# Histogram of Enhanced Image
plt.subplot(224)

for i, c in enumerate(colors):
    plt.hist(enhanced_img[:, :, i].ravel(), bins=256, range=[0, 256], color=c, alpha=0.5)

plt.title("Histogram of Enhanced Image")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
